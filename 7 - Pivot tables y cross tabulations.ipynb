```python
#Llamamos el set de datos publico 
Chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'
import pandas as pd 
df_chicago=pd.read_csv(Chicago_data)
df_chicago['CRASH_DATE'] = df_chicago['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce', utc=True))
report_1 = df_chicago.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR']).agg({'BEAT_OF_OCCURRENCE': sum})
```

```python
# Creando tabla pivote rapidamente sobre estos indices
report1 = pd.pivot_table(df_chicago, index=['LIGHTING_CONDITION', 'REPORT_TYPE'])
report1 = report1.reset_index()
report1.head()
```
