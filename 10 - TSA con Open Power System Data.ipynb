```python
# load time series dataset
import pandas as pd 
import numpy as np 
import seaborn as sns
```

### Dataset de la uni√≥n europea, en busqueda de migrar a energias limpias

```python
url='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/df_power.csv'
```

```python
df_power = pd.read_csv(url)
```

```python
df_power.columns
```

```python
# Reseteando la columna de date para que sea tipo datetime      
df_power['Date'] = pd.to_datetime(df_power['Date'])
```

```python
df_power = df_power.set_index('Date')
```

```python
# Al ser datatime tiene propiedades que nos permite separar por month, year, etc
df_power['year'] = df_power.index.year
```

```python
df_power.head()
```

```python
df_power['month'] = df_power.index.month
```

```python
df_power.head()
```

```python
df_power['day_name'] = df_power.index.day_name()
```

```python
import matplotlib.pyplot as plt
```

```python
df_power['Consumption'].plot(linewidth=0.5)
```

```python
name_cols = ['Consumption', 'Wind', 'Solar']

df_power[name_cols].plot(marker='.', alpha=0.5, linestyle='None',figsize=(14,6), subplots=True)
```

```python
df_power.loc['2016', 'Consumption'].plot()
```

```python
# Filtrando por anio y mes
df_power.loc['2016-12', 'Consumption'].plot()
```

```python
fig, axes = plt.subplots(3, 1,figsize=(8,7))

for name, ax in zip(['Consumption', 'Wind', 'Solar'], axes):
    sns.boxplot(data=df_power, x='month', y=name, ax=ax)
```
