```python
import pandas as pd
import altair as alt
#Llamamos el set de datos público 
Chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'

df_chicago = pd.read_csv(Chicago_data)
df_chicago['CRASH_DATE'] = df_chicago['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce', utc=True))
```

```python
df_chicago.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR'])
```

```python
# VAmos a utilizar aggregation
# Total de choques por condicion de luz y encodeado el tipo de reporte

report_1 = df_chicago.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR']).agg({'NUM_UNITS': sum})
report_1 = report_1.reset_index()
report_1
```

```python
# Total de choques por condicion de luz y encodeado el tipo de reporte

alt.Chart(report_1).mark_bar().title.encode(
    x='LIGHTING_CONDITION',
    y='NUM_UNITS',
    color='REPORT_TYPE',
    
).properties(width=400)
```

```python
## mas de 1 operacion con el aggregation

df_chicago.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR']).agg({'NUM_UNITS': ['sum', 'min', 'max']})
```
